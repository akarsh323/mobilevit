```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import keras.callbacks
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import MobileNetV3Small
from tensorflow.keras.metrics import Accuracy
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
from tensorflow.keras.callbacks import LearningRateScheduler

# Custom Self-Attention Layer
class SelfAttention(tf.keras.layers.Layer):
    def __init__(self, units):
        super(SelfAttention, self).__init__()
        self.units = units
        self.W_q = tf.keras.layers.Dense(units)
        self.W_k = tf.keras.layers.Dense(units)
        self.W_v = tf.keras.layers.Dense(units)
        self.attention = tf.keras.layers.Attention()

    def call(self, inputs):
        query = self.W_q(inputs)
        key = self.W_k(inputs)
        value = self.W_v(inputs)
        attention_output = self.attention([query, key, value])
        return attention_output

def inverted_residual_block(inputs, num_filters, strides=1, expansion_ratio=1):
    # Point-Wise Convolution
    x = tf.keras.layers.Conv2D(
        filters=expansion_ratio * inputs.shape[-1],
        kernel_size=1,
        padding="same",
        use_bias=False
    )(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("swish")(x)

    # Depth-Wise Convolution
    x = tf.keras.layers.DepthwiseConv2D(
        kernel_size=3,
        strides=strides,
        padding="same",
        use_bias=False
    )(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("swish")(x)

    # Point-Wise Convolution
    x = tf.keras.layers.Conv2D(
        filters=num_filters,
        kernel_size=1,
        padding="same",
        use_bias=False
    )(x)
    x = tf.keras.layers.BatchNormalization()(x)

    # Residual Connection
    if strides == 1 and (inputs.shape == x.shape):
        return tf.keras.layers.Add()([inputs, x])
    return x

def mlp(x, mlp_dim, dim, dropout_rate=0.1):
    x = tf.keras.layers.Dense(mlp_dim, activation="swish")(x)
    x = tf.keras.layers.Dropout(dropout_rate)(x)
    x = tf.keras.layers.Dense(dim)(x)
    x = tf.keras.layers.Dropout(dropout_rate)(x)
    return x

def transformer_encoder(x, num_heads, dim, mlp_dim):
    skip_1 = x
    x = tf.keras.layers.LayerNormalization()(x)
    x = tf.keras.layers.MultiHeadAttention(
        num_heads=num_heads, key_dim=dim
    )(x, x)
    x = tf.keras.layers.Add()([x, skip_1])

    skip_2 = x
    x = tf.keras.layers.LayerNormalization()(x)
    x = mlp(x, mlp_dim, dim)
    x = tf.keras.layers.Add()([x, skip_2])

    return x

def mobile_vit_block(inputs, num_filters, dim, patch_size=2, num_layers=1):
    B, H, W, C = inputs.shape

    # 3x3 conv
    x = tf.keras.layers.Conv2D(
        filters=C,
        kernel_size=3,
        padding="same",
        use_bias=False
    )(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("swish")(x)

    # 1x1 conv: d-dimension
    x = tf.keras.layers.Conv2D(
        filters=dim,
        kernel_size=1,
        padding="same",
        use_bias=False
    )(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("swish")(x)

    # Reshape x to flattened patches
    P = patch_size * patch_size
    N = int(H * W // P)
    x = tf.keras.layers.Reshape((P, N, dim))(x)

    # Transformer Encoder
    for _ in range(num_layers):
        x = transformer_encoder(x, 1, dim, dim * 2)

    # Reshape
    x = tf.keras.layers.Reshape((H, W, dim))(x)

    # 1x1 conv: C-dimension
    x = tf.keras.layers.Conv2D(
        filters=C,
        kernel_size=1,
        padding="same",
        use_bias=False
    )(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("swish")(x)

    # Concatenation
    x = tf.keras.layers.Concatenate()([x, inputs])

    # 3x3 conv
    x = tf.keras.layers.Conv2D(
        filters=num_filters,
        kernel_size=3,
        padding="same",
        use_bias=False
    )(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("swish")(x)

    return x

def MobileViT(input_shape, num_classes, pretrain=True):
    num_channels = [16, 32, 48, 48, 48, 64, 96, 80, 120, 96, 144, 384]
    dim = [96, 120, 44]
    expansion_ratio = 4

    # Input layer
    inputs = tf.keras.layers.Input(input_shape)

    # Stem
    x = tf.keras.layers.Conv2D(
        filters=num_channels[0],
        kernel_size=3,
        strides=2,
        padding="same",
        use_bias=False
    )(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("swish")(x)
    x = inverted_residual_block(x, num_channels[1], strides=1, expansion_ratio=expansion_ratio)

    # Stage 1
    x = inverted_residual_block(x, num_channels[2], strides=2, expansion_ratio=expansion_ratio)
    x = inverted_residual_block(x, num_channels[3], strides=1, expansion_ratio=expansion_ratio)
    x = inverted_residual_block(x, num_channels[4], strides=1, expansion_ratio=expansion_ratio)

    # Stage 2
    x = inverted_residual_block(x, num_channels[5], strides=2, expansion_ratio=expansion_ratio)
    x = mobile_vit_block(x, num_channels[6], dim[0], num_layers=1)

    # Stage 3
    x = inverted_residual_block(x, num_channels[7], strides=2, expansion_ratio=expansion_ratio)
    x = mobile_vit_block(x, num_channels[8], dim[1], num_layers=4)

    # Stage 4
    x = inverted_residual_block(x, num_channels[9], strides=2, expansion_ratio=expansion_ratio)
    x = mobile_vit_block(x, num_channels[10], dim[2], num_layers=3)
    x = tf.keras.layers.Conv2D(
        filters=num_channels[11],
        kernel_size=1,
        padding="same",
        use_bias=False
    )(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("swish")(x)

    # Classifier
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    # Use sigmoid activation for binary classification
    outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)

    model = tf.keras.models.Model(inputs, outputs)

    if pretrain:
        # Load pre-trained MobileNetV3 model without top (fully connected) layers
        base_model = MobileNetV3Small(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

        # Copy weights from MobileNetV3 to MobileViT
        for layer in base_model.layers:
            if layer.name in model.layers:
                model.get_layer(layer.name).set_weights(layer.get_weights())

    return model
# Function to plot training history
def plot_training_history(history, lr, optimizer):
    plt.figure(figsize=(12, 4))

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'Model Accuracy ({optimizer}, lr={lr})')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'], loc='upper left')

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'Model Loss ({optimizer}, lr={lr})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.show()

# Define your data directories
# Define your data directories
train_dir = '/home/chetak/workspace/Mini_Project_Dataset1/train'
test_dir = '/home/chetak/workspace/Mini_Project_Dataset1/test'
val_dir = '/home/chetak/workspace/Mini_Project_Dataset1/val'
# Load image paths and labels into a DataFrame
images = []
labels = []

for i in tqdm([train_dir, test_dir, val_dir]):
    for subfolder in os.listdir(i):
        subfolder_path = os.path.join(i, subfolder)
        if not os.path.isdir(subfolder_path):
            continue
        for image_filename in os.listdir(subfolder_path):
            image_path = os.path.join(subfolder_path, image_filename)
            images.append(image_path)
            labels.append(subfolder)

df = pd.DataFrame({'image': images, 'label': labels})
# Split the data
X_train, X_test1, y_train, y_test1 = train_test_split(df['image'], df['label'], test_size=0.3, random_state=42, shuffle=True, stratify=df['label'])
X_val, X_test, y_val, y_test = train_test_split(X_test1, y_test1, test_size=0.5, random_state=42, shuffle=True, stratify=y_test1)
df_train = pd.DataFrame({'image': X_train, 'label': y_train})
df_test = pd.DataFrame({'image': X_test, 'label': y_test})
df_val = pd.DataFrame({'image': X_val, 'label': y_val})
image_size = (256, 256)
batch_size = 32
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'

)
# Create data generators
train_generator = train_datagen.flow_from_dataframe(
    df_train,
    x_col='image',
    y_col='label',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',  # Adjusted for binary classification
    shuffle=True
)
test_generator = train_datagen.flow_from_dataframe(
    df_test,
    x_col='image',
    y_col='label',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',  # Adjusted for binary classification
    shuffle=False
)
val_generator = train_datagen.flow_from_dataframe(
    df_val,
    x_col='image',
    y_col='label',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',  # Adjusted for binary classification
    shuffle=True
)


import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def build_and_compile_model(input_shape, num_classes, learning_rate, optimizer):
    model = MobileViT(input_shape=input_shape, num_classes=num_classes, pretrain=True)

    if optimizer == 'adam':
        opt = Adam(learning_rate=learning_rate)
    else:
        raise ValueError("Invalid optimizer")

    model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])

    return model

# Define a grid of hyperparameters to search
param_grid = {
    'learning_rate': [1e-4, 1e-5, 1e-6]
}

# Iterate through all combinations with Adam optimizer
for lr in param_grid['learning_rate']:
    params = {'learning_rate': lr, 'optimizer': 'adam'}
    print(f"Training with learning rate {params['learning_rate']} and optimizer {params['optimizer']}")
    num_classes=2
    # Build and compile the model
    model = build_and_compile_model(input_shape=(*image_size, 3), num_classes=num_classes,
                                     learning_rate=params['learning_rate'], optimizer=params['optimizer'])

    # Add EarlyStopping
    early_stopping = EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)

    # Train the model
    history = model.fit(train_generator, epochs=500, validation_data=val_generator, callbacks=[early_stopping])

    # Save the model
    model.save(f'model_lr_{params["learning_rate"]}_NewDatasetmain.h5')
    # Evaluate on the test set
    loss, accuracy = model.evaluate(test_generator)
    print(f'Test accuracy: {accuracy}')

    # Plot training history
    plot_training_history(history, params['learning_rate'], params['optimizer'])

    # Plot confusion matrix
    y_true = test_generator.classes
    y_pred = (model.predict(test_generator) > 0.5).astype(int).flatten()

    cm = confusion_matrix(y_true, y_pred)
    display_labels = ['Class 0', 'Class 1']

    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=display_labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix ({params["optimizer"]}, lr={params["learning_rate"]})')
    plt.show()
```
